```python
import pandas as pd
import glob
df1 = pd.read_csv("ap_bus_details.csv")
df2 = pd.read_csv("assam_bus_details.csv")
df3 = pd.read_csv("chandigarh_bus_details.csv")
df4 = pd.read_csv("Haryana_bus_details.csv")
df5 = pd.read_csv("kadamba_bus_details.csv")
df6 = pd.read_csv("kerala_bus_details.csv")
df7 = pd.read_csv("Rajasthan_bus_details.csv")
df8 = pd.read_csv("west bengal_bus_details.csv")
df9 = pd.read_csv("Telangana_bus_details.csv")
df10 = pd.read_csv("up_bus_details.csv")

```

```python
filepaths_df=[df1,df2,df3,df4,df5,df6,df7,df8,df9,df10]
combined_df=pd.concat(filepaths_df,ignore_index=True)
combined_df
```

```python
combined_df.isnull().sum()
```

```python
combined_df= combined_df.dropna() #drop null values
```

```python
combined_df = combined_df.fillna(0) # Replace null values with 0
```

```python
combined_df['Price'] = combined_df['Price'].fillna(0)
combined_df
```

```python
# Remove Duplicates
combined_df = combined_df.drop_duplicates()
```

```python
# Remove the 'INR' and any spaces, then convert to float
combined_df['Price'] = combined_df['Price'].str.replace('INR', '').str.replace(' ', '').astype(float)

# Check the updated data types
print(combined_df.dtypes)
```

```python
combined_df.dtypes
```

```python
# Use .loc[] to avoid the SettingWithCopyWarning
combined_df.loc[:, ['Departing_Time', 'Reaching_Time']] = combined_df[['Departing_Time', 'Reaching_Time']].astype('string')

# Check the updated data types
print(combined_df.dtypes)
```

```python
combined_df.to_csv("combined_data.csv", index=False)
combined_df
```

```python
import mysql.connector
import pandas as pd

# Connect to MySQL
myconnection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="123456789",
    database="Redbus"
)
cursor = myconnection.cursor()

# Create the table only if it doesn't exist
cursor.execute("""
    CREATE TABLE IF NOT EXISTS bus_routes (
        route_name VARCHAR(255),
        route_link VARCHAR(255),
        bus_name VARCHAR(255),
        bus_type VARCHAR(255),
        departing_time VARCHAR(50),
        duration VARCHAR(50),
        reaching_time VARCHAR(50),
        star_rating FLOAT,
        price FLOAT,
        seat_availability VARCHAR(255)
    )
""")

# Load data and insert into the table
combined_df = pd.read_csv("combined_data.csv")

for i, row in combined_df.iterrows():
    try:
        cursor.execute("""
            INSERT INTO bus_routes (route_name, route_link, bus_name, bus_type, departing_time, 
                                    duration, reaching_time, star_rating, price, seat_availability)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """, (
            row['Route_Name'], 
            row['Route_Link'], 
            row['Bus_Name'], 
            row['Bus_Type'], 
            row['Departing_Time'], 
            row['Duration'], 
            row['Reaching_Time'], 
            row['Star_Rating'], 
            row['Price'], 
            row['Seat_Availability']
        ))
    except Exception as e:
        print(f"Error: {e}")

# Commit changes and close connection
myconnection.commit()
cursor.close()
myconnection.close()

print("Data inserted successfully into the MySQL database!")
```
